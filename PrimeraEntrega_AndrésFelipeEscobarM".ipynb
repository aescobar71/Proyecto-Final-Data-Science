Encodear Variables de tipo Object:

Primero se Transforman las variables categóricas que tienen un orden implícito ("Engine" y "FuelConsumtion"), indicando el orden en que deben quedar y usando la función for y luego entreno el algoritmo. Por último se imprimen los valores unicos después del encoder:

```python
from sklearn import preprocessing

encoder = ['Engine','FuelConsumtion']
hot_encoder = ['Brand','UsedOrNew','Transmission','DriveType','FuelType','BodyType','Doors','Seats']

order={'FuelConsumtion':['Consumo Bajo','Consumo Medio','Consumo Alto'],
       'Engine':['Motor Pequeño','Motor Estandar','Motor Grande']}

for col in encoder:

    # Initialize a LabelEncoder object
    label_encoder = preprocessing.LabelEncoder()

    # Fit the encoder to the unique values in the column
    print(col+'_order')
    label_encoder.fit(np.array(order[col]))

    # Transform the column using the encoder
    df[col] = label_encoder.transform(df[col])

    # Print the column name and the unique encoded values
    print(f"{col}: {df[col].unique()}")

df
```

Ahora transformamos las demás variables tipo Object:

```python
one_hot_encoder = preprocessing.OneHotEncoder(sparse=False)

# Ajustar y aplicar la codificación one-hot al DataFrame
data_encoded = pd.DataFrame(one_hot_encoder.fit_transform(df[hot_encoder]))

nombres_columnas = one_hot_encoder.get_feature_names_out(hot_encoder)
data_encoded.columns = nombres_columnas

data_encoded
```
