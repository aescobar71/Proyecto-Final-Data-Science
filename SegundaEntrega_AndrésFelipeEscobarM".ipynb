```python
from sklearn import preprocessing

encoder = ['Engine', 'FuelConsumtion']
hot_encoder = ['DriveType', 'FuelType', 'BodyType', 'Brand', 'State']

order={'FuelConsumtion':['Consumo Bajo', 'Consumo Medio', 'Consumo Alto', 'Consumo Muy Alto'],
       'Engine':['Motor Pequeño', 'Motor Estandar', 'Motor Grande', 'Motor Muy Grande']}

for col in encoder:

    # Initialize a LabelEncoder object
    label_encoder = preprocessing.LabelEncoder()

    # Fit the encoder to the unique values in the column
    print(col+'_order')
    label_encoder.fit(np.array(order[col]))

    # Transform the column using the encoder
    df[col] = label_encoder.transform(df[col])

    # Print the column name and the unique encoded values
    print(f"{col}: {df[col].unique()}")

df
```

### **Árbol de Decisión (Decision Tree Regressor)**

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(random_state=1234, max_depth=4, min_samples_leaf=3, min_samples_split=8)
dtree.fit(X_train, y_train)
```

Ahora genero las métricas:

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_percentage_error
import math

y_pred = dtree.predict(X_test)

mae = metrics.mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)

print('MAE is {}'.format(mae))
print('MAPE is {}'.format(mape))
print('MSE is {}'.format(mse))
print('R2 score is {}'.format(r2))
print('RMSE score is {}'.format(rmse))
```

Las predicciones del modelo se desvían aproximadamente 12,346 unds o un 37.1% del valor real. El valor de R2 de 0.5139 indica que aproximadamente el 51.4% de la variabilidad en los datos reales es explicada por este modelo. Por tanto, el modelo de árbol de decisión tiene una capacidad razonable para predecir los valores, pero también tiene un margen significativo de error, por lo cual es adecuado probar otro modelo diferente.
